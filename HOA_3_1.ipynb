## Part 2: Scatterplot Graphs and Correlatable Variables

Step 1: The pandas describe() method.


```python
brainFrame.describe()
```

Step 2: Scatterplot graphs

a. Load the required modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

b. Seperate the data

```python
menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

c. Plot the graphs.

```python
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.show()
%matplotlib inline
```

Observation:  It shows here the datas on the table to be able to create a scatterplot to have a quick visual of the commonalities of the values of PIQ,FSIQ, VIQ.

## PART 3: Calculating Correlation with Python

Step 1: Calculate correlation against brainFrame

```python
brainFrame.corr(method = "pearson")
```


* Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filled with 1s? Is that a coincidence? Explain.

  When correlation uses the variable would be always equal to itself so it is not a coincidence.

* Still looking at the correlation table above, notice that the values are mirrored; values below the 1
diagonal have a mirrored counterpart above the 1 diagonal. Is that a coincidence? Explain

  The correlation table above displays a mirrored value below the 1 is not a coincidence because the value of each variable should be always equal to it self that's why it is mirrored.

```python
womenDf.corr(method = "pearson")
```

```python
menDf.corr(method = "pearson")
```

Observation: In this part, we used a method of correlation which is pearson to able to create a linear correlation table between two variables. By observing the tables you can see now the difference between the data on women and men.
