```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_wine = pd.read_csv('/content/Wine.csv')
```


The problem in this dataset that we are trying to solve is to predict the quality of Vinho Verde, a type of wine, based on the various physicochemical attributes provided in the dataframe.

```python
df_wine.head()
```

```python
df_wine.info()
```

```python
df_wine.isnull().sum()
```

```python
df_wine.describe()
```

```python
#Independent Variable
y = pd.get_dummies(df_wine["quality"])
y
```

```python
#Dependent Variable
x = df_wine.drop(["quality", "Id"], axis=1)
x
```

**SPLITTING DATA**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
x_train
```

```python
y_train
```

**MULTILAYER PERCEPTRON**

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

#For number of hidden layers
model.add(Dense(5, activation='sigmoid'))
#For number of  output nodes
model.add(Dense(6, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',metrics=["accuracy"])
```

```python
#Number of iteration
train_history = model.fit(x_train, y_train, batch_size=25, epochs=50, validation_data=(x_test,y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose=False)

print('Test score: ', score[0])    #Loss on test
print('Test accuracy: ', score[1])
```

#Model Evaluation

The loss test, quantifying the model's overall performance, stands at 1.1167978048324585. The higher the loss test, the poorer the model's performance. The model's accuracy is only 0.5218340754508972, indicating that it's not performing at a high level. This low accuracy can be attributed to the inadequacy of the network structure, as it forms the foundation for achieving higher accuracy. This accuracy level implies that the model correctly predicts the outcome for only about half of the test instances, suggesting room for improvement in creating a model.
